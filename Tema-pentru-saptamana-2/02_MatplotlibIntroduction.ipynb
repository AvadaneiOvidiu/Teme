```python
x = np.linspace(0, 1, 10)
random_sample = np.random.rand(1000)
```

```python
threshes = np.dstack((x[:-1], x[1:])).reshape(x.size - 1, 2)
print(threshes)
```

```python
counts = np.array([random_sample[(random_sample >= t[0]) & (random_sample < t[1])].size for t in threshes])
print(counts)
```

```python
plt.plot(x[:-1], counts, label = 'uniform_distribution')

plt.xlabel('x')
plt.ylabel('Empiric count')

plt.show()
```

```python
figure, ax = plt.subplots()

ax.plot(x[:-1], counts)

ax.set(xlabel='x', ylabel='count',
       title='Empiric Counts')
ax.grid()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
loaded=np.loadtxt('numbers.txt',delimiter=',')
print(loaded)
```

```python
A=np.reshape(loaded,(1,np.product(loaded.shape)))
print(A)
B=A.reshape(4,6)
print(B)
```

```python
C=B.reshape(1,24)
print(C)
```

```python
minimum=C.min()
maximum=C.max()
print(C.size)
v=np.arange(minimum,maximum+1)
counts1=[]
numbers=[]
frequencies=[]
for i in range(0,v.size):
    c=0
    for j in range(0,C.size):
        if C[0][j]==i:
            c=c+1
    if c!=0:
        numbers.append(i)
        frequencies.append(c)
densities=[]
for i in range(0,len(frequencies)):
    densities.append(frequencies[i]/C.size)
plt.plot(numbers[:],densities)
```

```python
cummulative_distribution=[]
for i in range(0,len(densities)):
    c=0;
    for j in range(0,i+1):
        c+=densities[j]
    cummulative_distribution.append(c)
plt.plot(numbers,cummulative_distribution)
```
